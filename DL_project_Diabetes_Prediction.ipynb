```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

Dataset

```python
# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
# printing the first 5 rows of the dataset
diabetes_dataset.head()
```

```python
# number of rows and Columns in this dataset
diabetes_dataset.shape
```

```python
# getting the statistical measures of the data
diabetes_dataset.describe()
```

0 --> Non-Diabetic

1 --> Diabetic

```python
diabetes_dataset['Outcome'].value_counts()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# separating the data and labels
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training

```python
# Build the ANN model
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(X_train.shape[1],)))  # Input layer with 16 neurons
model.add(Dense(8, activation='relu'))  # Hidden layer with 8 neurons
model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

Model evaluation phase

```python
# Train the model
model.fit(X_train, Y_train, epochs=100, batch_size=10)

# Evaluate the model on training data
train_loss, train_accuracy = model.evaluate(X_train, Y_train)
print('Training Accuracy: {:.2f}%'.format(train_accuracy * 100))

# Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(X_test, Y_test)
print('Test Accuracy: {:.2f}%'.format(test_accuracy * 100))

# Make predictions on test data
Y_pred = (model.predict(X_test) > 0.5).astype("int32")

# Print classification report for more metrics
print(classification_report(Y_test, Y_pred))

# Example input data for prediction
input_data = (5, 166, 72, 19, 175, 25.8, 0.587, 51)

# Changing the input_data to numpy array and scaling it
input_data_as_numpy_array = np.asarray(input_data).reshape(1, -1)
input_data_scaled = scaler.transform(input_data_as_numpy_array)

# Make prediction with the ANN model
prediction = (model.predict(input_data_scaled) > 0.5).astype("int32")
print(prediction)

if (prediction[0][0] == 0):
    print('The person is not diabetic')
else:
    print('The person is diabetic')

# Save the trained model using pickle (for Keras models use save method)
model.save('trained_ann_model.h5')

# Load the saved model to make predictions again (optional)
from tensorflow.keras.models import load_model

loaded_model = load_model('trained_ann_model.h5')

# Make prediction with loaded model for verification
prediction_loaded = (loaded_model.predict(input_data_scaled) > 0.5).astype("int32")
print(prediction_loaded)

if (prediction_loaded[0][0] == 0):
    print('The person is not diabetic')
else:
    print('The person is diabetic')
```
